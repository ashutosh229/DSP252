## 3. Problem: Write a Python program to solve the quadratic equation of the form ax^2 + bx + c = 0 for given a, b, and c. The program should return the two roots (if they exist) or a message if the equation has no real roots.

````python
import math

def solve_quadratic(a, b, c):
    discriminant = b**2 - 4 * a * c
    if discriminant > 0:
       #Implement your code here....
       root1 = (-b + math.sqrt(discriminant))/(2*a)
       root2 = (-b - math.sqrt(discriminant))/(2*a)
       return [root1,root2]

    elif discriminant == 0:
        #Implement your code here....
        root = -b/(2*a)
        return [root]

    else:
        return "No real roots"


a, b, c = 1, -3, 2
result = solve_quadratic(a, b, c)
if result == "No real roots":
  print(result)
elif len(result)==1:
  print(f"The roots of the equation are real and equal and the root is {result[0]}")
else:
  print(f"The roots of the equation are real and unequal and the roots are {result[0]} and {result[1]}")
````

## 4. Problem: Write a Python program to generate all Pythagorean triplets (a, b, c) where a^2 + b^2 = c^2 for a <= b <= c <= n.

````python
def pythagorean_triplets(n):
    triplets = []
    for a in range(1, n + 1):
        for b in range(a, n + 1):
            for c in range(b, n + 1):
                if a**2 + b**2 == c**2:
                    triplets.append((a, b, c))
    return triplets


n = 20
triplets = pythagorean_triplets(n)
print(f"Pythagorean triplets up to {n}: {triplets}")
````

## 5. Problem: Write a Python program to find all Armstrong numbers between two numbers n1 and n2.

````python
def is_armstrong(n):
    #Implement your code here....
    nString = str(n)
    sum = 0
    for digit in nString:
      component = int(digit)**len(nString)
      sum = sum + component
    return sum == n


def armstrong_numbers_in_range(n1, n2):
    #Implement your code here....
    armstrong_numbers = []
    for num in range(n1+1,n2):
      if is_armstrong(num):
        armstrong_numbers.append(num)
    return armstrong_numbers

n1, n2 = 100, 999
print(f"Armstrong numbers between {n1} and {n2}: {armstrong_numbers_in_range(n1, n2)}")
````
