## Calculate the mean,median,mode,standard deviation and variance for the given dataset:

## [5,10,15,20,25,30,35]

```python
import numpy as np
import statistics as stats
```

```python
data=np.array([5,10,15,20,25,30,35])
```

```python
mean = stats.mean(data)
median = stats.median(data)
mode = stats.mode (data)
std = stats.stdev(data)
var = stats.variance(data)
```

```python
print(f"The mean of the given data is: {mean}")
print(f"The median of the given data is: {median}")
print(f"The mode of the given data is: {mode}")
print(f"The standard deviation of the given data is: {std}")
print(f"The variance of the given data is: {var}")
```

```python
import numpy as np
import statistics as stats
```

```python
data
```

```python
reshapedData = np.reshape(data,(4,5))
```

```python
reshapedData
```

```python
reshapedData2 = np.reshape(reshapedData,(20))
```

```python
reshapedData2
```

```python
print(f"The mean of the given data is: {mean}")
print(f"The median of the given data is: {median}")
print(f"The mode of the given data is: {mode}")
print(f"The standard deviation of the given data is: {std}")
print(f"The variance of the given data is: {var}")
```
